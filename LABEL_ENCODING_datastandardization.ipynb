Label Encoding:
converting the labels into numeric form

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("/content/data (1).csv")
df.head()
```

```python
cancer=df.drop('Unnamed: 32',axis=1)
```

```python
cancer.head()
```

```python
cancer.isnull().sum()
```

```python
cancer['diagnosis'].value_counts()
```

```python
#load the label encoder function
label=LabelEncoder()
```

```python
labels=label.fit_transform(cancer.diagnosis)
```

```python
#appending the labels to the data frame
cancer['target']=labels
cancer.head()
```

```python
cancer['target'].value_counts()
```
